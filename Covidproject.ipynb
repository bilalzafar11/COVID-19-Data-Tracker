```python
for country in countries:
    sns.lineplot(data=covid[covid['location']==country],label=country,x='date',y='new_cases')
    plt.title('New COVID-19 Cases per Day')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.show()
```

```python
latest_data = covid[covid['date'] == covid['date'].max()]
sns.lineplot(data=latest_data, x='location', y='total_deaths')
plt.title('Total Deaths by Country (Latest)')
plt.xlabel('Country')
plt.ylabel('Total Deaths')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=covid, x='date', y='people_vaccinated', hue='location')
plt.title('Vaccination Progress Over Time')
plt.ylabel('People Vaccinated')
plt.show()
```

```python
corr_covid = covid[['total_cases', 'total_deaths', 'people_vaccinated']].corr()
sns.heatmap(corr_covid, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
